## Import data

```python
# import necessary packages
import numpy as np
import MySQLdb
import pandas as pd
from bokeh import plotting, io, models, layouts
```

```python
# set output mode
io.reset_output()
io.output_notebook()
```

```python
# database credentials
dsn_database = "ercm_kc135"
dsn_hostname = "localhost"
dsn_port = 3306
dsn_uid = "root"
dsn_pwd = "root"

# create the database connection
conn = MySQLdb.connect(host = dsn_hostname, port = dsn_port, user = dsn_uid, passwd = dsn_pwd, db = dsn_database)
```

```python
# create an empty dataframe
df = []

# select last five years of sortie history
query = "SELECT * FROM ercm_kc135.three_month_fh_collector"

# create dataframe from sortie history
df = pd.read_sql_query(query, conn)
```

```python
# check the data load
df.head()
```

## plot the data

```python
# generate axis labels
cnames = list(df.columns.values)
xlabel = cnames[0].replace("_", " ")
ylabel = cnames[1].replace("_", " ")

print(xlabel)
print(ylabel)
```

```python
# create initial plot
plot = plotting.figure(title = 'Average Flight Hours for Three Month Rolling Window',
                       x_axis_type="datetime", 
                       x_axis_label = xlabel,
                       y_axis_label = ylabel,
                       width=1000)

plot.line(x = df[cnames[0]],
          y = df[cnames[1]])

plot.xaxis.formatter = models.DatetimeTickFormatter(
        hours=["%Y-%m"],
        days=["%Y-%m"],
        months=["%Y-%m"],
        years=["%Y-%m"],
    )
```

```python
io.output_notebook()
io.show(plot)
```
